```python
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

from sklearn.linear_model import LinearRegression
```

## 2. Creating a Simple Dataset

Let's create a tiny dataset:

| Hours Studied (X) | Exam Score (y) |
|-------------------|----------------|
|         1         |       40       |
|         2         |       50       |
|         3         |       60       |
|         4         |       70       |
|         5         |       80       |

We will pretend:

- These are the **known** data points.
- Then we'll ask:  
  _"If someone studies 6 or 7 hours, what will be their score?"_


```python

X = np.array([[1],
              [2],
              [3],
              [4],
              [5]]) 

y = np.array([40, 50, 60, 70, 80]) 
df = pd.DataFrame({
    "Hours Studied": X.ravel(),
    "Exam Score": y
})

df
```

```python
plt.scatter(X, y)

plt.xlabel("Hours Studied (X)")
plt.ylabel("Exam Score (y)")
plt.title("Hours Studied vs Exam Score")
plt.show()
```

```python
model = LinearRegression()
model.fit(X, y)

b = model.coef_[0]
a = model.intercept_

print("Slope (b):", b)
print("Intercept (a):", a)
print(f"So the model learned: y = {b:.2f} * x + {a:.2f}")
```

```python
y_pred = model.predict(X)

comparison_df = pd.DataFrame({
    "Hours Studied (X)": X.ravel(),
    "Actual Score (y)": y,
    "Predicted Score (ŷ)": y_pred
})

comparison_df
```

```python
X_new = np.array([[6],
                  [7],
                  [8]])   # Only X, y is unknown

y_new_pred = model.predict(X_new)

new_pred_df = pd.DataFrame({
    "Hours Studied (X_new)": X_new.ravel(),
    "Predicted Exam Score (ŷ_new)": y_new_pred
})

new_pred_df
```
