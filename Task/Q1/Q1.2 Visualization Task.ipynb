# Task 2 – Data Loading & Basic Visualizations

In this task, we learn how to:
- Load datasets using pandas  
- Understand feature distributions  
- Plot categorical and numerical data  
- Create scatter plots  
- Visualize correlation using heatmaps  

These steps help beginners understand how data looks before training any machine learning model.


```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

## Loading a Sample Dataset

We load a dataset using `pandas.read_csv()`.

If you have your own dataset, replace `'file.csv'` with the file name.
For demonstration, we use the built-in **Iris dataset** from seaborn.


```python
df = sns.load_dataset("iris")
df.head()
```

## Plot Distribution of a Numerical Feature (Histogram)

A histogram shows how a numerical value is spread out.

For example:
- Are most values small?
- Are they large?
- Do they form a bell curve?

Here we plot the distribution of **sepal_length**.

```python
plt.figure(figsize=(6,4))
plt.hist(df["sepal_length"], bins=20, edgecolor='black')
plt.title("Distribution of Sepal Length")
plt.xlabel("Sepal Length")
plt.ylabel("Frequency")
plt.show()
```

## Plot Count of Samples in Each Category

When working with categorical data (like species in Iris),  
we use a **bar chart** or **count plot** to see how many samples belong to each class.


```python
plt.figure(figsize=(6,4))
sns.countplot(x='species', data=df)
plt.title("Species Count")
plt.xlabel("Species")
plt.ylabel("Count")
plt.show()
```

## Scatter Plot – Understanding Relationships

A scatter plot helps us see:
- How two features relate  
- If they form clusters  
- If one increases with the other  

Here we compare **sepal_length** vs **petal_length**.

```python
plt.figure(figsize=(6,4))
sns.scatterplot(x='sepal_length', y='petal_length', hue='species', data=df)
plt.title("Sepal vs Petal Length")
plt.show()
```

## Correlation Heatmap

A heatmap shows how strongly features relate to each other.

- +1 → Strong positive relation
- -1 → Strong negative relation
- 0 → No relation

This is useful for feature selection.


```python
plt.figure(figsize=(8,6))

numeric_df = df.select_dtypes(include='number')   # only numeric columns
corr_matrix = numeric_df.corr()                   # correlation of numeric features

sns.heatmap(corr_matrix, annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```

or use 

```python
plt.figure(figsize=(8,6))
sns.heatmap(df.corr(numeric_only=True), annot=True, cmap='coolwarm')
plt.title("Correlation Heatmap")
plt.show()
```
