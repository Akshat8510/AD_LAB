```python
import pandas as pd

print("\n=== TEST CASE 1: MISSING VALUE HANDLING (MEAN METHOD) ===\n")

data = {
    'CustomerID': [1, 2, 3, 4],
    'Age': [25, None, 30, None],
    'Income': [50000, 60000, None, 55000],
    'Gender': ['Male', 'Female', None, 'Female']
}

df = pd.DataFrame(data)

print("Original Data:\n")
print(df.to_string(index=True))

# Check missing values
print("\nMissing Values Count:\n")
print(df.isna().sum().to_string())

# Fill missing values
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])

print("\nData After Applying Mean/Mode Method:\n")
print(df.to_string(index=True))

print("\n=== TEST CASE 1 COMPLETED ✅ ===")
import pandas as pd
from sklearn.preprocessing import LabelEncoder, MinMaxScaler

# -------------------------------
# TEST CASE 2: INCORRECT DATA
# -------------------------------
print("\n=== TEST CASE 2: INCORRECT DATA (MISSING & WRONG VALUES) ===\n")

data = {
    'Customer ID': [10, 11, 12, 13, 14],
    'Age': [28, None, 35, 22, None],
    'Gender': ['Male', 'Female', None, 'Female', 'Male'],
    'Income': [70000, 60000, None, 45000, 52000],
    'City': ['Urban', 'Rural', 'Urban', 'Rural', 'Urban'],
    'Subscription Status': ['Subscribed', None, 'Subscribed', 'Not Subscribed', None]
}

df = pd.DataFrame(data)

print("Original Incorrect Data:\n")
print(df.to_string(index=True))

# -------------------------------
# Missing Value Check
# -------------------------------
print("\nMissing Values Count:\n")
print(df.isna().sum().to_string())

# -------------------------------
# Handle Missing Values
# -------------------------------
df['Age'] = df['Age'].fillna(df['Age'].mean())
df['Income'] = df['Income'].fillna(df['Income'].mean())
df['Gender'] = df['Gender'].fillna(df['Gender'].mode()[0])
df['Subscription Status'] = df['Subscription Status'].fillna(
    df['Subscription Status'].mode()[0]
)

print("\nData After Handling Missing Values:\n")
print(df.to_string(index=True))

# -------------------------------
# Encoding
# -------------------------------
le = LabelEncoder()
df['Gender_Encoded'] = le.fit_transform(df['Gender'])
df['Subscription_Encoded'] = le.fit_transform(df['Subscription Status'])

df = pd.get_dummies(df, columns=['City'])

print("\nData After Encoding:\n")
print(df.to_string(index=True))

# -------------------------------
# Feature Scaling
# -------------------------------
scaler = MinMaxScaler()
df[['Age_Scaled', 'Income_Scaled']] = scaler.fit_transform(df[['Age', 'Income']])

print("\nFinal Processed Data (After Scaling):\n")
print(df.to_string(index=True))

print("\n=== TEST CASE 2 COMPLETED SUCCESSFULLY ✅ ===")
```

```python
# ==========================================================
# TASK 3: SIMPLE LINEAR REGRESSION ON HOUSING PRICE DATASET
# (Single Cell – Google Colab Ready)
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ----------------------------------------------------------
# TEST CASE 1: LOAD & INSPECT DATASET
# ----------------------------------------------------------
print("\n=== TEST CASE 1: LOAD & INSPECT DATASET ===\n")

df = pd.read_csv("housing_price_dataset.csv")

print("First 5 Rows:\n")
print(df.head().to_string())

print("\nDataset Info:\n")
df.info()

print("\nMissing Values Per Column:\n")
print(df.isna().sum().to_string())

# ----------------------------------------------------------
# TEST CASE 2: SELECT FEATURE & TARGET
# ----------------------------------------------------------
print("\n=== TEST CASE 2: FEATURE & TARGET SELECTION ===\n")

X = df[["Avg. Area Income"]]   # Feature
y = df["Price"]               # Target

print("Feature (X) Preview:\n")
print(X.head().to_string())

print("\nTarget (y) Preview:\n")
print(y.head().to_string())

# ----------------------------------------------------------
# TEST CASE 3: TRAIN SIMPLE LINEAR REGRESSION MODEL
# ----------------------------------------------------------
print("\n=== TEST CASE 3: MODEL TRAINING ===\n")

model = LinearRegression()
model.fit(X, y)

print(f"Slope (Coefficient): {model.coef_[0]}")
print(f"Intercept: {model.intercept_}")

# ----------------------------------------------------------
# TEST CASE 4: PREDICTIONS & COMPARISON
# ----------------------------------------------------------
print("\n=== TEST CASE 4: PREDICTIONS ON TRAINING DATA ===\n")

y_pred = model.predict(X)

comparison_df = pd.DataFrame({
    "Avg. Area Income": X["Avg. Area Income"].head(10),
    "Actual Price": y.head(10),
    "Predicted Price": y_pred[:10]
})

print(comparison_df.to_string(index=False))

# ----------------------------------------------------------
# TEST CASE 5: MODEL EVALUATION
# ----------------------------------------------------------
print("\n=== TEST CASE 5: MODEL EVALUATION ===\n")

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print(f"Mean Squared Error (MSE): {mse}")
print(f"R² Score: {r2}")

# ----------------------------------------------------------
# TEST CASE 6: VISUALIZE REGRESSION LINE
# ----------------------------------------------------------
print("\n=== TEST CASE 6: REGRESSION LINE VISUALIZATION ===\n")

plt.figure(figsize=(7,5))
plt.scatter(X["Avg. Area Income"], y, label="Actual Data")

sorted_idx = np.argsort(X["Avg. Area Income"].values)
plt.plot(
    X["Avg. Area Income"].values[sorted_idx],
    y_pred[sorted_idx],
    color="red",
    label="Regression Line"
)

plt.xlabel("Avg. Area Income")
plt.ylabel("Price")
plt.title("Simple Linear Regression: Avg. Area Income vs Price")
plt.legend()
plt.show()

# ----------------------------------------------------------
# TEST CASE 7: PREDICT NEW VALUES (NO WARNING VERSION)
# ----------------------------------------------------------
print("\n=== TEST CASE 7: PREDICT NEW VALUES ===\n")

X_new = pd.DataFrame({
    "Avg. Area Income": [20000, 50000, 90000]
})

y_new_pred = model.predict(X_new)

new_pred_df = pd.DataFrame({
    "Avg. Area Income": X_new["Avg. Area Income"],
    "Predicted Price": y_new_pred
})

print(new_pred_df.to_string(index=False))

print("\n=== TASK 3 COMPLETED SUCCESSFULLY ✅ ===")
```

```python
# ==========================================================
# TASK 4: MULTIPLE LINEAR REGRESSION ON HOUSING PRICE DATASET
# (SRM / Lab Style – Single Cell – With Graph)
# ==========================================================

import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# ----------------------------------------------------------
# TEST CASE 1: LOAD & INSPECT DATASET
# ----------------------------------------------------------
print("\n=== TEST CASE 1: LOAD & INSPECT DATASET ===\n")

df = pd.read_csv("housing_price_dataset.csv")

print("First 5 Rows:\n")
print(df.head().to_string())

print("\nMissing Values Per Column:\n")
print(df.isna().sum().to_string())

# ----------------------------------------------------------
# TEST CASE 2: FEATURE & TARGET SELECTION
# ----------------------------------------------------------
print("\n=== TEST CASE 2: FEATURE & TARGET SELECTION ===\n")

X = df[
    [
        "Avg. Area Income",
        "Avg. Area House Age",
        "Avg. Area Number of Rooms",
        "Avg. Area Number of Bedrooms",
        "Area Population"
    ]
]

y = df["Price"]

print("X Shape:", X.shape)
print("y Shape:", y.shape)

# ----------------------------------------------------------
# TEST CASE 3: TRAIN MODEL
# ----------------------------------------------------------
print("\n=== TEST CASE 3: MODEL TRAINING ===\n")

model = LinearRegression()
model.fit(X, y)

print("Intercept (a):", model.intercept_)
print("\nCoefficients:")
for col, coef in zip(X.columns, model.coef_):
    print(f"{col}: {coef}")

# ----------------------------------------------------------
# TEST CASE 4: PREDICTION
# ----------------------------------------------------------
print("\n=== TEST CASE 4: ACTUAL vs PREDICTED (FIRST 10) ===\n")

y_pred = model.predict(X)

comparison_df = pd.DataFrame({
    "Actual Price": y[:10],
    "Predicted Price": y_pred[:10]
})

print(comparison_df.to_string(index=False))

# ----------------------------------------------------------
# TEST CASE 5: MODEL EVALUATION
# ----------------------------------------------------------
print("\n=== TEST CASE 5: MODEL EVALUATION ===\n")

mse = mean_squared_error(y, y_pred)
r2 = r2_score(y, y_pred)

print("Mean Squared Error (MSE):", mse)
print("R² Score:", r2)

# ----------------------------------------------------------
# TEST CASE 6: GRAPH (SRM-ACCEPTED)
# Actual vs Predicted Price
# ----------------------------------------------------------
print("\n=== TEST CASE 6: ACTUAL vs PREDICTED PRICE GRAPH ===\n")

plt.figure(figsize=(7,5))
plt.scatter(y, y_pred, alpha=0.6)
plt.plot([y.min(), y.max()], [y.min(), y.max()], 'r--')  # ideal fit line
plt.xlabel("Actual Price")
plt.ylabel("Predicted Price")
plt.title("Multiple Linear Regression: Actual vs Predicted Price")
plt.show()

# ----------------------------------------------------------
# TEST CASE 7: PREDICTION FOR NEW DATA
# ----------------------------------------------------------
print("\n=== TEST CASE 7: PREDICTION FOR NEW HOUSE DATA ===\n")

X_new = pd.DataFrame({
    "Avg. Area Income": [50000, 75000],
    "Avg. Area House Age": [5, 10],
    "Avg. Area Number of Rooms": [6, 8],
    "Avg. Area Number of Bedrooms": [3, 4],
    "Area Population": [30000, 45000]
})

y_new_pred = model.predict(X_new)

new_pred_df = pd.DataFrame({
    "Avg. Area Income": X_new["Avg. Area Income"],
    "Avg. Area House Age": X_new["Avg. Area House Age"],
    "Rooms": X_new["Avg. Area Number of Rooms"],
    "Bedrooms": X_new["Avg. Area Number of Bedrooms"],
    "Population": X_new["Area Population"],
    "Predicted Price": y_new_pred
})

print(new_pred_df.to_string(index=False))

print("\n=== TASK 4 COMPLETED SUCCESSFULLY ✅ ===")
```
